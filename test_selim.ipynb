```python
import numpy as np  
import pandas as pd  
import datetime as dt 
import matplotlib.pyplot as plt 
```

```python
path="C:/Users/user/Desktop/work_selim"
df_ventes = pd.read_csv("ventes.csv",sep="\t")
df_ventes.describe() 
```

```python
df_articles= pd.read_csv("articles.csv",sep="\t")
df_articles.describe()
```

```python
df_clients= pd.read_csv("clients.csv",sep="\t")
df_clients.describe()
```

```python
#merge
merged=pd.merge(df_ventes,df_articles,how="inner",on="Article_ID")
merged
```

```python
df_clients.columns=['Client_ID','Effectif_Client', 'Categorie_Client']
```

```python
ventes_all=pd.merge(merged,df_clients,how='inner',on="Client_ID")
```

```python
filtered_ventes=ventes_all.dropna(subset=['Montant','Quantite_vendue'],axis=0,how='any')
```

```python
filtered_ventes
```

```python
mean=filtered_ventes['Montant'].mean()
```

```python
std=filtered_ventes['Montant'].std()
threshold=mean+3*std
```

```python
#j=[x for x in range(0,filtered_ventes.shape[0]) if filtered_ventes['Montant'][x]>100]


Index_label = filtered_ventes[filtered_ventes['Montant']<=threshold]
```

```python
Index_label
```

```python
Index_label["Montant_unitaire"]=Index_label["Montant"]/Index_label["Quantite_vendue"]
```
